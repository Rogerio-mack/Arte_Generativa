```python
#@markdown Basic imports
import matplotlib.pyplot as plt
import random
import math
from samila import GenerativeImage, Projection
```

## Exemplo 1 **Random generate.**

 Você pode simplesmente gerar uma figura em que o `samila` escolhe aleatoriamente as funções empregadas.



 

```python
g1 = GenerativeImage()
g1.generate()                    
g1.plot(color="#3E137D",bgcolor="orange")  # para a notação RBG de cores consulte aqui https://www.rapidtables.com/web/color/RGB_Color.html
plt.show()
```

## Exemplo 2 **Escolha suas funções.**

Mais interessante é você escolher as funções que deseja empregar como geradoras das imagens. As funções são funções matemáticas de duas variáveis $f(x,y) \rightarrow z $ e representadas por funções em `Python`. Elas podem ou não conter valores aleatórios. 

Aqui empregamos as funções:

$$ f_1(x,y) = x^2 - sin(y^2) + |x-y|$$
$$ f_2(x,y) = y^3 - cos(y^2) + 2x + \text{noise} $$


```python
def f1(x, y):
    result = x**2  - math.sin(y**2) + abs(y-x)                      # função sem componente aleatória
    return result

def f2(x, y):
    result = random.uniform(-1,1) * y**3 - math.cos(x**2) + 2*x     # função com componente aleatória
    return result

g2 = GenerativeImage(f1, f2)
g2.generate(seed=610510)                  # seed, para fixa a imagem gerada. Retire esse parâmetro para diferentes imagens.
g2.plot(color='blue',bgcolor='pink')
plt.show()
g2.seed
```

## Exemplo 3 **Projeções**

Você pode empregar diferentes projeções da sua figura. Uma projeção é, por exemplo, uma transformação da sua função para um outro tipo de coordenadas diferente das coordenadas cartesianas. Experimente as projeções  `POLAR`, `HAMMER` e `LAMBERT`, ou procure por mais projeções na documentação do pacote. 

```python
g2 = GenerativeImage(f1, f2)
g2.generate(seed=610510)                  
g2.plot(color='blue',bgcolor='pink',projection=Projection.POLAR)
plt.show()
g2.seed
```
